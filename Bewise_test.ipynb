```python
import pandas as pd
```

```python
test_data = pd.read_csv('test_data.csv')
test_data.head()
```

```python
test_data['greeting'] = False
greetings = ['здравствуйте', 'Здравствуйте', 'добрый день', 'Добрый день', 'доброе утро', 'Добррое утро', 
             'добрый вечер', 'Добрый вечер']
for i in range(480):
    test_data['greeting'][i] = any(elem in str(test_data['text'][i]) for elem in greetings)
```

```python
# a. Извлекаем реплики с приветствием – где менеджер поздоровался:
test_data[test_data['greeting'] == True][test_data['role'] == 'manager']['text']
```

```python
test_data['introduce'] = False
introduces = ['зовут', 'мое имя']
for i in range(480):
    test_data['introduce'][i] = any(elem in str(test_data['text'][i]) for elem in introduces)
test_data
```

```python
# b. Извлекаем реплики, где менеджер представил себя:
test_data[test_data['introduce'] == True][test_data['role'] == 'manager']['text']
```

```python
test_data['name'] = ''
for i in range(480):
    if test_data['introduce'][i] == True:
        l = str(test_data['text'][i]).split()
        for j in range(len(l)):
            if l[j] == 'зовут':
                test_data['name'][i] = l[j+1]
test_data
```

```python
# c. Извлекаем имя менеджера:
test_data[test_data['name'] != '']['name']
```

```python
test_data['company'] = ''
for i in range(480):
    if test_data['introduce'][i] == True:
        l = str(test_data['text'][i]).split()
        for j in range(len(l)):
            if l[j] == 'компания':
                test_data['company'][i] = l[j+1]
test_data
```

```python
# d. Извлекаем название компании:
test_data[test_data['company'] != '']['company']
```

```python
test_data['bye'] = False
goodbye = ['до свидания', 'До свидания', 'хорошего вечера', 'Хорошего вечера', 'всего хорошего', 'Всего хорошего', 
           'хорошего дня', 'Хорошего дня', 'всего доброго', 'Всего доброго']
for i in range(480):
    test_data['bye'][i] = any(elem in str(test_data['text'][i]) for elem in goodbye)
test_data
```

```python
# e. Извлекаем реплики, где менеджер попрощался:
test_data[test_data['bye'] == True][test_data['role'] == 'manager']['text']
```

```python
# f. Проверяем требование к менеджеру: «В каждом диалоге обязательно необходимо поздороваться и попрощаться с клиентом»
df2 = test_data[test_data['role'] == 'manager'][['dlg_id', 'greeting', 'bye']]
df2 = df2.groupby('dlg_id').agg({'greeting': 'max', 'bye': 'max'})
df2['fulfillment'] = df2['greeting'] & df2['bye']
df2
# Таким образом, видим, что требование выполняется в диалогах 0, 1 и 3.
```
