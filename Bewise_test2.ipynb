```python
from spacy.lang.ru import Russian
```

```python
nlp = Russian()
doc = nlp("Съешь ещё этих мягких французских булок, анна, да выпей чаю.")

for token in doc:
    print(token.text, token.lemma_, token.pos_, token.is_stop)
```

```python
%%timeit
test_data['name'] = ''
docs = list(nlp.pipe(test_data['text']))
for i in range(len(docs)):
    for token in docs[i]:
        if token.pos_ == 'PROPN':
            test_data.loc[[i], ['name']] += token.text + ' '
test_data
```

```python
%%timeit
test_data['name'] = ''
for i in range(len(test_data)):
    if test_data['introduce'][i] == True:
        text = nlp(test_data['text'][i])
        for token in doc:
            if token.pos_ == 'PROPN':
                test_data.loc[[i], ['name']] += token.text + ' '

test_data
```

```python
# c. Извлекаем имя менеджера:
test_data[test_data['name'] != '']['name']
```

```python
test_data['company'] = ''
for i in range(len(test_data)):
    if test_data['introduce'][i] == True:
        l = str(test_data.loc[[i], ['text']]).split()
        for j in range(len(l)):
            if l[j] == 'компания':
                test_data.loc[[i], ['company']] = l[j+1]
test_data
```

```python
# d. Извлекаем название компании:
test_data[test_data['company'] != '']['company']
```

```python
goodbye = ['до свидания','хорошего вечера', 'всего хорошего', 'хорошего дня', 'всего доброго']

test_data['bye'] = False

for bye in goodbye:
    test_data['bye'] |= test_data['text'].str.lower().str.contains(bye)
    
test_data
```

```python
# e. Извлекаем реплики, где менеджер попрощался:

filtered_test_data = test_data[test_data['bye']]
filtered_test_data = filtered_test_data[filtered_test_data['role'] == 'manager']
#другой вариант:
# filtered_test_data = test_data[(test_data['bye']) & (test_data['role'] == 'manager')]
filtered_test_data['text']
```

```python
# f. Проверяем требование к менеджеру: «В каждом диалоге обязательно необходимо поздороваться и попрощаться с клиентом»
df2 = test_data[test_data['role'] == 'manager'][['dlg_id', 'greeting', 'bye']]
df2 = df2.groupby('dlg_id').agg({'greeting': 'max', 'bye': 'max'})
df2['fulfillment'] = df2['greeting'] & df2['bye']
df2
# Таким образом, видим, что требование выполняется в диалогах 0, 1 и 3.
```
